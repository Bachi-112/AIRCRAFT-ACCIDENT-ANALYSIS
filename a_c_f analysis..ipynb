# AIRCRAFT CRASH AND FATALITY ANALYSIS.
## This is an exploratory data analysis on an aircraft accident dataset, from 1908 to 2009.
## The following are the Objectives for the data analysis:
   *  To find out how the frequency of aviation accidents has changed over the last century.
   *  Whether flights are getting safer in terms of survivability or not.
   *  Identifying which aircraft types have the most incident records.
   *  Comparing the safety records of different operators.
   *  Which regions or countries have the highest density of reported crashes?
   

## Defining the variables involved:
   * index: the index of the row
   * Date: the date of the incident
   * Time: the time of the incident
   * Location: the location of the incident
   * Operator: the operator of the aircraft
   * Flight #: the flight number of the aircraft
   * Route: the route of the aircraft
   * Type: the type of aircraft
   * Registration: the registration of the aircraft
   * cn/In: the construction number/serial number of the aircraft
   * Aboard: the number of people on board the aircraft
   * Fatalities: the number of people who died in the incident
   * Ground: the number of people on the ground killed in the incident
   * Summary: a summary of the incident

# DATA INSPECTION & CLEANING

```python
acf.isnull().sum()

# The columns with null values are :
# Time 
# Location
# Operator 
# Flight #
# Route
# Type 
# Registration
# cn/In
# Aboard
# Ground 
# Summary 

```

```python
acf["Time"].dtype
#Checking the data type for values under Time
```

```python
acf['Time'] = pd.to_datetime(acf['Time'], errors='coerce')
#Changing the datatype from Str to Datetime
```

```python
#Dropping the Time column, it has no significant effect on the data, and the insights i have to draw
acf = acf.drop (columns = ["Time"])
```

```python
#Dropping the index column. The data is already indexed.
acf = acf.drop (columns = ["index"])
```

```python
acf['Date'] = pd.to_datetime(acf['Date'])
acf['Year'] = acf['Date'].dt.year
acf['Month'] = acf['Date'].dt.month
acf['Day'] = acf['Date'].dt.day
#This converts the values under Date into Datetime format, and separates the Date column into "YEAR" "MONTH" AND "DATE", new columns.
```

```python
acf[[ 'Year', 'Month', 'Day']].describe()
#Now i have three new separate columns
```

```python
acf = acf.drop(columns=['Date'])
acf = acf.drop(columns=['Registration'])
acf = acf.drop(columns=['Route'])
acf = acf.drop(columns=['Flight #'])
acf = acf.drop(columns=['cn/In'])
acf = acf.drop(columns=['Ground'])
acf = acf.drop(columns=['Month'])
acf = acf.drop(columns=['Day'])
acf = acf.drop(columns=['Summary'])
#I drop the date column since i already have the values in separate new columns(3)
```

```python
## FILLING THE NULL VALUES UNDER CERTAIN COLUMNS:
acf["Operator"] = acf["Operator"].fillna("Anonymous")
acf["Location"] = acf["Location"].fillna("UNKNOWN")
#These columns were filled because they affect a huge percenteage of the data
```

```python
#DROPPING NULLS FOR CERTAIN COLUMNS
acf.dropna(subset=["Type"],inplace=True)
acf.dropna(subset=["Fatalities"],inplace=True)
acf.dropna(subset=["Aboard"],inplace=True)
#theyre dropped because they do not affect a significant portion of the data
```

```python
acf.head()
```

```python
acf.to_csv("cleaned_Airplane_Crashes_and_Fatalities_Since_1908.csv", index = False)
```

# DATA ANALYSIS


```python
sbs.heatmap(acf.corr(numeric_only=True), annot=True, cmap='coolwarm')
#From the heatmap below, the conditions strongly tied to fatalities are number of people aboard the airplane and Year
```

```python
plt.rcParams.update({
    'font.family': 'DejaVu Sans',          
    'font.weight': 'bold',           
    'axes.titlesize': 18,            
    'axes.labelsize': 14,            
    'xtick.labelsize': 12,           
    'ytick.labelsize': 12})           
plt.style.use('dark_background')
#Setting the default graph appearance
```

```python
#Graph number one,, showing fatality frequency against time in years
plt.title("FATALITY FREQUENCY AGAINST TIME")
plt.Figure(figsize = (20,20))
sbs.lineplot(data= acf, y = "Fatalities" , x = "Year",color = "#FFBF00" )

plt.show()
```

```python
acf ["fatality_rate"] = acf["Fatalities"]/acf["Aboard"]
#An arithmetic computation that calculates the fatality rate as a fraction.     This is a new column im creating
acf = acf.dropna(subset=['fatality_rate'])
```

```python
## A regplot that shows how fatality rate has changed over time
plt.title("Fatality rate over the years")
plt.Figure(figsize=(10,15))
sbs.regplot(data = acf, x = "Year", y = "fatality_rate",color = "#FFBF00",line_kws={'linewidth': 3}, scatter_kws= {'alpha':0.3} )
plt.grid(color='white', linestyle='--', linewidth=0.5, alpha=0.3)
plt.show()
```

```python
# Picking the top 5 most frequently appearing aircraft types and plotting a bargraph of fatalities against aircraft type 
top_5_types = acf['Type'].value_counts().head(5)
top_5_df = top_5_types.reset_index()
top_5_df.columns = ['Aircraft Type', 'Fatalities']
plt.Figure(figsize=(20,10))
plt.title("Aircraft Types' Fatality Magnitudes")
sbs.barplot(data = top_5_df, x = "Fatalities", y = "Aircraft Type", color = "#FFBF00")
plt.show()
```

```python
## Filtering the Operator column to the top five most appearing and plotting a bar graph showing fatality magnitude for each operator
top_5_data = acf['Operator'].value_counts().head(5)
top_5_acf = top_5_data.reset_index()
top_5_acf.columns = ['Category', 'Count']
plt.title("Aircraft Operators' Fatality Magnitudes")
sbs.barplot(data = top_5_acf, x = "Count", y = "Category", color = "#FFBF00")
plt.xlabel("Number of fatalities")
plt.ylabel("Operator")
plt.show()
```

```python
top_5_loc = acf['Location'].value_counts().head(5)#creating a variable that represents the five most frequent locations
top_5_l = top_5_loc.reset_index()# Move the index labels into a standard column so they can be accessed as data for plotting.
top_5_l.columns = ['Place', 'Fatalities']#Naming the standard columns
plt.title("Accident locality against Fatality Magnitudes")
sbs.barplot(data = top_5_l, x = "Fatalities", y = "Place", color = "#FFBF00")
plt.xlabel("Number of Fatalities")
plt.ylabel("Place")
plt.show()
```

# FINDINGS FROM THE ABOVE ANALYSIS:
   #### 1. The frequency of aviation accidents has increased over time.This is probably tied to increase in aircraft transport users
   #### 2. Flights are indeed getting safer in terms of survivability, possibly due to technical and technological advancements in the constantly changing field
   #### 3. The aircraft model with highest fatalities is the Douglas DC-3.This is most likely owed to the fact that it was heavily used in the first world war.
   #### 4. The operator with the most recorded fatalities is Aeroflot,a Russian operator.Coupled with the participation of Russia in World war I and the country's rugged terrain, this explains the significantly high fatality count.
   #### 5.Most accidents occur in unknown locations. This may be associated with loss of direction that comes with aircraft accidents.
